```python
import pandas as pd
import numpy as np
import pickle
from utilities.utils import *
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
get_job_skills = pd.read_csv("./data/sea_job_skill_full_2021_v1.3.1.csv", low_memory=False)
```

```python
get_job_skills.info()
```

```python
job_w_skill_df = get_job_skills[get_job_skills.skill!='no skill extracted']
job_wo_skill_df = get_job_skills[get_job_skills.skill=='no skill extracted']
```

```python
print(f"Total number of skills extracted: {job_w_skill_df.shape[0]} skills, from {len(job_w_skill_df.jobId.unique())} unique jobs.")
print(f"There are {len(job_wo_skill_df.jobId.unique())} unique jobs with no skills extracted.")
```

```python
total_job_cnt = len(get_job_skills.jobId.unique())
total_job_cnt
```

```python
assert total_job_cnt == len(job_wo_skill_df.jobId.unique()) + len(job_w_skill_df.jobId.unique())
```

```python
job_tsc_app_df = job_w_skill_df[job_w_skill_df.skill_form.isin(["TSC","App/Tools"])]
```

```python
job_tsc_app_df['skill'] = job_tsc_app_df.skill.apply(lambda x: x.lower().strip())
```

```python
job_tsc_app_df.drop(columns=['skill_form'], inplace=True)
```

```python
job_tsc_app_df.head(1)
```

```python
job_skill_dict = df_to_dict(job_tsc_app_df.jobId, job_tsc_app_df.skill)
```

```python
job_skill_dict
```

```python
job_skill_list = []
for _id, skill in job_skill_dict.items():
    job_skill_list += [skill]
```

```python
assert len(job_skill_list) == len(job_tsc_app_df.jobId.unique())
print(len(job_skill_list))
```

```python
print(job_skill_list[:2])
```

```python
#Pickling
with open("./model/job_skill_corpus", "wb") as fp:
    pickle.dump(job_skill_list, fp)
```
